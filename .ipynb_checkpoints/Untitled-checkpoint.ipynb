```python
df = df.drop(columns=['X1'])
```

```python
X = df.loc[:,['age', 'gender', 'height', 'weight', 'steps', 'hear_rate',
       'calories', 'distance', 'entropy_heart', 'entropy_setps',
       'resting_heart', 'corr_heart_steps', 'norm_heart', 'intensity_karvonen',
       'sd_norm_heart', 'steps_times_distance',
       'device_apple watch', 'device_fitbit']]
Y = df.loc[:,'activity']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
Y.dtypes
```

```python
 X_train
```

```python
y_train.head(5)
```

```python
 from sklearn.ensemble import RandomForestClassifier 
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy 
```

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)  # Define K-Fold cross-validation with 5 folds

```

```python
cv_scores = cross_val_score(rf_classifier , X,Y, cv=kfold, scoring='accuracy')
```

```python
print("Cross-Validation Scores:", cv_scores)
print("Mean Accuracy:", cv_scores.mean())
print("Standard Deviation of Accuracy:", cv_scores.std())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(X_train, y_train)
y_pred = logistic_regression.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svm_classifier = SVC()
svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)
 
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier()
 
knn_classifier.fit(X_train, y_train)
 
y_pred = knn_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
decision_tree_classifier = DecisionTreeClassifier()

decision_tree_classifier.fit(X_train, y_train)

y_pred = decision_tree_classifier.predict(X_test)
 
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
naive_bayes_classifier = GaussianNB()

naive_bayes_classifier.fit(X_train, y_train)
 
y_pred = naive_bayes_classifier.predict(X_test)
 
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
